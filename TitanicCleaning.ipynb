```python
import pandas as pd
```

```python
import numpy as np
```

```python
titanicData = pd.read_csv("E:/szkola/3_semestr/SIWB/titanic/TitanicMess.tsv", sep='\t')
```

```python
pd.set_option('display.max_rows', None)
```

```python
titanicData
```

```python
#First of all, lets see NaNs appearance
titanicData.isna().any()
```

```python
titanicData = titanicData.fillna(0)
# Found some NaNs so replacing with default 0
```

```python
#Embarked seems to contain wrong values
titanicData['Embarked'].unique()
```

```python
#Let's fix the wrong values
titanicData['Embarked'] = titanicData['Embarked'].replace(['Qe'],'Q')
titanicData['Embarked'] = titanicData['Embarked'].replace(['So'],'S')
titanicData['Embarked'] = titanicData['Embarked'].replace(['Co'],'C')

titanicData['Embarked'].unique()
```

```python
#Next thing is Ship column All passangers has the same value for this column
titanicData['ship'].unique()
```

```python
#Let's delete that column
del titanicData['ship']
```

```python
#Sex column values has some typos
titanicData['Sex'].unique()
```

```python
titanicData['Sex'] = titanicData['Sex'].replace(['malef'],'male')
titanicData['Sex'] = titanicData['Sex'].replace(['mal'],'male')
titanicData['Sex'] = titanicData['Sex'].replace(['fem'],'female')
titanicData['Sex'] = titanicData['Sex'].replace(['femmale'],'female')
titanicData['Sex'].unique()
```

```python
#Let's check results
titanicData['Sex'].unique()
```

```python
#When it comes to Fare column there are few things to consider.
#1. Floats should has "." instead of "," as  a decimal point delimiter
titanicData['Fare'] = titanicData['Fare'].str.replace(',','.')
#2. There seems to be a value "15.9a" which is not a number. let's fix that 
titanicData['Fare'] = titanicData['Fare'].str.replace('15.9a','15.9')
#3. And finally we can easly round the decimal point to 2
titanicData['Fare'] = titanicData['Fare'].astype(float).round(2)
```

```python
titanicData['Fare']
```

```python
titanicData['Age'].unique()
```

```python
titanicData['Age'] = titanicData['Age'].str.replace('.9','0.9')
```

```python
#Lets replace delimiter 
titanicData['Age'] = titanicData['Age'].str.replace(',','.')
```

```python
titanicData['Age'].unique()
```

```python
#decimal values are unnecessary lets round it down
titanicData['Age'].astype(float).apply(np.floor)
```

```python
# There is a value 4435 which seems to be way to big for a human being age.
# The oldest person alive is Jeanne Calment and is 122 years old now. 
# So ill optimisticly round the age up to 130 and down to 0
titanicData['Age'] = titanicData['Age'].astype(float).apply(lambda x: x if x > 0 else 0)
```

```python
titanicData['Age'] = titanicData['Age'].astype(int).apply(lambda x: x if x <= 130 else 0)
```
